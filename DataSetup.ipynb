### Movie Table

```python
import numpy as np
import json
```

```python
df_movie = pd.read_csv("movies_metadata.csv")
```

```python
def give_genre(gjson):
    genreJ = gjson.replace("'",'"').replace('None','"NA"')
    try:
        t = json.loads(genreJ)[0]['name']
        return t
    except Exception as e:
        print(i," : ",e)
        return np.nan
```

```python
df_movie.columns
```

```python
### remove dates not in correct format 
df_movie = df_movie[pd.to_datetime(df_movie['release_date'], errors='coerce').notna()]
```

```python
df_movie.shape
```

```python
df_movie["runtime"]= df_movie["runtime"].fillna(0).astype(int)
```

```python
df_movie["title"].fillna("ABC",inplace=True)
```

```python
Movie_table = []
for i in range(df_movie.shape[0]):
    if i in [19730]:
        continue
    temp = {}
    temp["id"] = int(i+1)
    temp["title"]= df_movie.iloc[i].title
    temp["release_date"]= df_movie.iloc[i].release_date
    temp["runtime"]= int(df_movie.iloc[i].runtime)
    temp["genre"] = give_genre(df_movie.iloc[i].genres)
    temp["adder_id"] = "user"+str(random.randint(1,4000)) ### only first 4000 users add a moviw.
    Movie_table.append(temp)
    
```

```python
xx = pd.DataFrame(Movie_table)
xx.to_csv("MovieTable.csv", index=False,header=False)
```

### Cast Table

```python
import ast
```

```python
df_links = pd.read_csv('credits.csv')
```

```python
castjson = df_links.cast.iloc[5]
```

```python
cast_data = []
for i in range(df_links.shape[0]):
    castjson = df_links.cast.iloc[i]
    cast_dict = ast.literal_eval(castjson)
    
    for i in cast_dict:
        temp = {}
        temp['id'] = i['id']
        temp['Name'] = i['name']
        temp['gender'] = i['gender']
        cast_data.append(temp)

```

```python
cx = pd.DataFrame(cast_data)
cx = cx.drop_duplicates(["id"]).set_index("id").sort_index()
```

```python
cx["Name"].isna().sum()
cx.shape
```

```python
cx.to_csv("CastTable.csv", index=True,header=False)
```

```python
role_data = []
for i in range(df_links.shape[0]):
    castjson = df_links.cast.iloc[i]
    cast_dict = ast.literal_eval(castjson)
    temp2 = {}
    temp2["movie_id"] = i+1
    if i+1 not in mids:
        continue
    for cd in cast_dict:
        temp = temp2.copy()
        temp['cast_id'] = cd['id']
        temp['role'] = cd['character']
        if temp['cast_id'] not in cids:
            continue
        role_data.append(temp)
```

```python
import numpy as np
```

```python
import string

def generate_random_string():
    # Example: Generate a random string of length between 5 and 10
    length = random.randint(1, 200)
    characters = string.ascii_letters + string.digits  # You can include other characters as needed
    return ''.join(random.choice(characters) for _ in range(length))



print(f"Random String: {generate_random_string()}")
```

```python
df_review = pd.read_csv("ratings.csv")
```

```python
# df_review["xx"] =
df_review["userId"] = "user"+df_review["userId"].astype(str)
```

```python
def random_desc():
    x = random.randint(1,7)
    if x==3:
        return generate_random_string()
    else:
        return np.nan
```

```python
df_review.shape[0]
```

```python
desc = []
for i in range(26024289):
    desc.append(random_desc())
```

```python
desc[0:10]
```

```python
dfdesc = pd.Series(desc)
```

```python
df_review["rating"] = df_review["rating"].apply(np.ceil)
```

```python
df_review["Desc"] = dfdesc
```

```python
df_review.drop(['timestamp'], axis=1,inplace=True)
```

```python
m = pd.DataFrame(mids,columns=["mid"])
```

```python
df_x = df_review.merge(m,left_on="movieId",right_on="mid")
```

```python
df_x.drop("mid",axis=1,inplace=True)
```

```python
df_x["rating"] = df_x["rating"].astype(int)
```
